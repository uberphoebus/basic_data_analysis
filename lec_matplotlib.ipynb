```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.title('Line Graph')
plt.show()
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
```

```python
plt.plot(x, y)
plt.title('꺾은 선 그래프')
```

```python
matplotlib.rcParams['axes.unicode_minus'] = False
plt.plot([-1, 0, 1], [-5, -1, 2])
```

# 2. 축

```python
plt.plot(x, y)
plt.title('꺾은 선 그래프', fontdict={'family':'HYGungSo-Bold', 'size':20})
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red', loc='right') # left, center, right
plt.ylabel('Y축', color='#00aa00', loc='top') # top, center, bottom
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```

# 3. 범례(legend)

```python
plt.plot(x, y)
plt.legend('W', loc='lower right')
plt.show()
```

```python
plt.plot(x, y)
plt.legend('W', loc=(0.5, 0.5))
plt.show()
```

# 4. 스타일

```python
plt.plot(x, y, linewidth=5, marker='o')
plt.show()
```

```python
plt.plot(x, y, marker='v', linestyle='None', markersize=10)
plt.show()
```

```python
plt.plot(
    x, y, marker='X', markersize=20,
    markerfacecolor='yellow', markeredgecolor='red'
)
plt.show()
```

```python
plt.plot(x, y, linestyle=':')
plt.show()
```

```python
plt.plot(x, y, linestyle='--')
plt.show()
```

```python
plt.plot(x, y, linestyle='-.')
plt.show()
```

```python
plt.plot(x, y, color='pink')
plt.show()
```

```python
plt.plot(x, y, color='b')
plt.show()
```

```python
plt.plot(x, y, color='#ff0000')
plt.show()
```

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```
